## Which genre combinations are the most profitable?


```python
df_csv_1 = pd.read_csv('Project Code/tmdb_5000_credits.csv')
#movie id, title, case, crew
```

```python
df_csv_2 = pd.read_csv('Project Code/tmdb_5000_movies.csv')
#budget, genres, homepage, id', keywords', original_language', original_title', overview', popularity', 
#production_companies', production_countries', release_date', revenue', runtime', spoken_languages', status',
#tagline', title', vote_average', vote_count'
```

```python
merged_inner = pd.merge(left = df_2, right = df_csv_2, left_on = 'title', right_on = 'title')
df_genre_profit = merged_inner[['title', 'genres_x', 'revenue']].sort_values(by = 'revenue', ascending = False)

genre_list_step_1 = []
genre_list_step_2 = []

for i in range(50):
    genre_list_step_1.append(df_genre_profit['genres_x'].values[i].strip('][').split(', '))
        
for x in genre_list_step_1:
    genre_list_step_2.append(tuple(x))
    
dictionary_1 = Counter(genre_list_step_2)
final_dictionary = {}
sort_dictionary = sorted(dictionary_1.items(), key = lambda x:x[1], reverse = True)
for item1, item2 in sort_dictionary:
    final_dictionary[item1] = item2

gen_x = list(final_dictionary.values())[:5]
gen_y = [str(item) for item in list(final_dictionary.keys())][0:5]

fig_genre_profit, ax_gp = plt.subplots(figsize = (10, 5), dpi = 80)

ax_gp.set_title("Most Common Genre Combination Amongst The 100 Most Profitable Films")
ax_gp.hlines(y=gen_y, xmin=0, xmax=4, color='gray', alpha=0.7, linewidth=1, linestyles='dashdot')
ax_gp.scatter(y=gen_y, x=gen_x, s=75, color='firebrick', alpha=0.7)
ax_gp.set_xlabel("Repetitions");
```

## Which Individual Genres Are The Most Profitable?

```python
df_genre_profit_2 = df_genre_profit
genre_dict = {}
#df_genre_profit_2['genres_x'].values[0].strip('][').split(', ')
for i in range(127):
    for item in df_genre_profit_2['genres_x'].values[i].strip('][').split(', '):
        if item in list(genre_dict.keys()):
            genre_dict[item] += df_genre_profit_2['revenue'][i]
        else:
            genre_dict[item] = df_genre_profit_2['revenue'][i]


```

```python
fig_2, ax_2 = plt.subplots(figsize = (15, 5))
x2 = list(genre_dict.keys())[:10]
y2 = list(genre_dict.values())[:10]
y3 = []

for num in y2:
    y3.append(round(num/1000000000))


ax_2.vlines(x=x2, ymin=0, ymax=11.0, color='firebrick', alpha=0.7, linewidth=2)
ax_2.scatter(x=x2, y=y3, s=75, color='firebrick', alpha=0.7)
ax_2.set_title("Cumulative Revenue per Genre in Billions")
ax_2.set_xlabel("Genres")
ax_2.set_ylabel("Approximate Revenue (Billions)");
```

## Relationship Between Budget & Ratings AND Budget & Vote Count

## Directors Who Have Directed More Then One Movie
## That Are Considered Highly Profitable & Very Popular

```python
df_act = df
df_act = df_act.dropna(subset = ['director', 'original air date'])
df_act['year'] = df_act['year'].apply(int).values
df_act = df_act[['title', 'director', 'year', 'Gross_Income']]

director_list = []
#index_1 = df['director'][i].index('_')
#index_2 = (df['director'][i][(index_1 + 1):].index('_')) + index_1
#index_3 = (df['director'][i][(index_2 + 1):].index('_')) + index_2
#index_4 = (df['director'][i][(index_3 + 1):].index('_')) + index_3
for item in df_act['director'].values:
    if item.count('_') == 2:
        index_1 = item.index('_')
        index_2 = (item[(index_1 + 1):].index('_')) + index_1
        director_list.append(item[(index_1 + 1):(index_2 + 1)])
    elif item.count('_') == 4:
        index_1 = item.index('_')
        index_2 = (item[(index_1 + 1):].index('_')) + index_1
        index_3 = (item[(index_2 + 1):].index('_')) + index_2
        index_4 = (item[(index_3 + 1):].index('_')) + index_3
        director_list += [item[(index_1 + 1):(index_2 + 1)], item[(index_3 + 1):(index_4 + 1)]]

for item in director_list:
    if item == '':
        director_list.remove(item)

director_dict = Counter(director_list)

for item in list(director_dict.keys()):
    if director_dict[item] <= 1:
        director_dict.pop(item)
        
director_dict
```

```python
df_pop = merged_inner[['title', 'director', 'vote_count', 'year']]
df_pop = df_pop.sort_values(['vote_count'], ascending = False)
df_pop['year'] = df_pop['year'].apply(int).values
df_pop = df.dropna(subset = ['director'])

director_pop_list = []
for item in df_pop['director'].values:
    if item.count('_') == 2:
        index_1 = item.index('_')
        index_2 = (item[(index_1 + 1):].index('_')) + index_1
        director_pop_list.append(item[(index_1 + 1):(index_2 + 1)])
    elif item.count('_') == 4:
        index_1 = item.index('_')
        index_2 = (item[(index_1 + 1):].index('_')) + index_1
        index_3 = (item[(index_2 + 1):].index('_')) + index_2
        index_4 = (item[(index_3 + 1):].index('_')) + index_3
        director_pop_list += [item[(index_1 + 1):(index_2 + 1)], item[(index_3 + 1):(index_4 + 1)]]

for item in director_pop_list:
    if item ==  '':
        director_pop_list.remove(item)
        
exp_list = director_list + director_pop_list
omg_dict = Counter(exp_list)

for item in list(omg_dict.keys()):
    if omg_dict[item] <= 5:
        omg_dict.pop(item)
omg_dict
```

```python
dir_fig, dir_ax = plt.subplots(figsize = (25, 5))

dir_x = list(omg_dict.keys())
dir_y = list(omg_dict.values())

dir_ax.bar(dir_x, dir_y)
```
